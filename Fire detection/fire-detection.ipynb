```python
import tensorflow as tf
import cv2
import numpy as np
from tensorflow.keras.preprocessing import image
from playsound import playsound
```

```python
fire_model = tf.keras.models.load_model("fire_detection.hdf5")
```

```python
fire_model.summary()
```

```python
cap = cv2.VideoCapture("Fire Incident.mp4")
alert=[]
alarm =False

while cap.isOpened():
    ret, frame = cap.read()
    
#     img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    cv2.imwrite("Pics/dsd.jpg",frame)
    
    file = r"Pics/dsd.jpg"
    
    
    img = image.load_img(file, target_size=(224,224))
    img_array = np.asarray(img)
    img_array = img_array.reshape(1,img_array.shape[0],img_array.shape[1],img_array.shape[2])
    classes = fire_model.predict(img_array)
    classes
    lab=np.argmax(classes)
    label = "False" if lab==1 else "True"
    
    alert.append(label)
    
    if len(alert)==8:
        
        print(alert)
        
        if alert.count("True")>4:
            print("Alarmed")
            cv2.putText(frame, "ALARMING", (80, 300), cv2.FONT_HERSHEY_SIMPLEX, 2, (255,155,144), 3)
#             playsound(audio_path)
        
        
        alert=[]   
        
    
    color = (0,255,0) if label=="False" else (0,0,255)
#     Make detections 
    cv2.putText(frame, "Fire:  ", (10, 70), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 255), 2)
    cv2.putText(frame, (str(label)), (80, 70), cv2.FONT_HERSHEY_SIMPLEX, 0.9, color, 2)
    cv2.imshow("frame",frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
